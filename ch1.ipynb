```python
# The first line creates a smaller list because we are normalizing the text by lowercasing each word before we cast the list of words into a set. If we do it the other way around and create a set of all the words in the text first, and then make a list of them all in lowercase, then there will be duplicate words if a word was used with different capitalization at different points in the text.
print(len(sorted(set(w.lower() for w in text1))))
print(len(sorted(w.lower() for w in set(text1))))
# This should hold true for all the other texts as well. 
list = [text2, text3, text4, text5, text6, text7, text8, text9]
for i in range(len(list)):
    first = len(sorted(set(w.lower() for w in text1)))
    second = len(sorted(w.lower() for w in set(text1)))
    print("{}: {}".format(list[i], first < second))
```
