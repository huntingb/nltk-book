```python
from nltk.corpus import swadesh
fr2en = swadesh.entries(["fr", "en"])
translate = dict(fr2en)
de2en = swadesh.entries(["de", "en"])
es2en = swadesh.entries(["es", "en"])
translate.update(dict(de2en))
translate.update(dict(es2en))
print(translate)
# The way the object is now, it's just a dictionary storing one-way translation relationships. This isn't very space efficient because we have duplicate entries for every different language we are translating to english..
translate = dict()
def add_entries(tuplist):
    for i in range(len(tuplist)):
        if tuplist[i][1] not in translate:
            translate[tuplist[i][1]] = [tuplist[i][0]]
        else:
            translate[tuplist[i][1]].append(tuplist[i][0])
add_entries(fr2en)
add_entries(de2en)
add_entries(es2en)
print(translate)
def lookup(word):
    for i in range(len(translate.values())):
        if word in list(translate.values())[i]:
            return list(translate.keys())[i]
print(lookup("noir"))
print(lookup("sucio"))
```

```python
def supergloss(s):
    gloss = s.definition()
    for i in range(len(s.hyponyms())):
        gloss += s.hyponyms()[i].definition() + " "
    for i in range(len(s.hypernyms())):
        gloss += s.hypernyms()[i].definition() + " "
    return gloss

print(supergloss(wn.synsets("tree")[0]))
```

```python
from nltk.book import *
from nltk.corpus import stopwords

def most_common_not_stop(text):
    stop = stopwords.words("english")
    common = nltk.FreqDist(text).most_common()
    ret = []
    i = 0
    while len(ret) < 50 and i < len(common):
        word = common[i][0]
        if word.lower() not in stop:
            ret.append(word)
        i += 1
    return ret
           
print(most_common_not_stop(text1))
```

```python
def most_common_bigrams(text):
    stop = stopwords.words("english")
    bigrams = nltk.bigrams(text)
    common = nltk.FreqDist(bigrams).most_common()
    ret = []
    i = 0
    while len(ret) < 50 and i < len(common):
        bigram = common[i][0]
        if not (bigram[0] in stop or bigram[1] in stop):
            ret.append(bigram)
        i += 1
    return ret

print(most_common_bigrams(text1))
```

```python
def syllables(word):
    vowels = ["A", "E", "I", "O", "U"]
    prondict = nltk.corpus.cmudict.dict()
    if word not in prondict:
        return
    prons = prondict[word]
    possible = []
    for pron in prons:
        syllables = 0
        if pron[0][0] in vowels:
            syllables += 1
        for i in range(len(pron) - 1):
            if pron[i][0] not in vowels and pron[i+1][0] in vowels:
                syllables += 1
        possible.append(syllables)
    return possible

print(syllables("a"))
print(syllables("apple"))
print(syllables("beautiful"))
print(syllables("apparently"))
print(syllables("capitalism"))
print(syllables("syzygy"))
```

```python
import nltk
from matplotlib import pyplot as plt
from nltk.book import *

def zipf():
    texts = [text1, text2, text3, text4, text5, text6, text7, text8, text9]
    for text in texts:
        fdist = nltk.FreqDist(text).most_common()
        ranks = list(range(len(fdist)))
        counts = [fdist[i][1] for i in range(len(fdist))]
        plt.plot(ranks, counts)
    plt.xlabel("Word rank")
    plt.ylabel("Word count")
    plt.xscale("log")
    plt.yscale("log")
    plt.title("Zipf's law")
    plt.legend(texts, loc="upper right", fontsize="xx-small")
    plt.show()
    
zipf()
```

```python
def generate_hybrid(cfd1, cfd2, word, num=15):
    for i in range(num):
        print(word, end=" ")
        random_cfd = random.choice([cfd1, cfd2])
        other_cfd = cfd1 if random_cfd == cfd2 else cfd2
        random_next_words = random_cfd[word]
        if len(random_next_words):
            word = random.choice(list(random_next_words.keys()))
        else:
            other_next_words = other_cfd[word]
            word = random.choice(list(other_next_words.keys()))

moby = nltk.bigrams(text1)
cfd2 = nltk.ConditionalFreqDist(moby)

generate_hybrid(cfd, cfd2, "God", 100)
print("\n")
generate_hybrid(cfd, cfd2, "Whale", 100)
```

```python
from nltk.corpus import udhr
langs = [l for l in udhr.fileids() if l.endswith("Latin1")]
def word_in_langs(word):
    ret = []
    for lang in langs:
        if word.lower() in [w.lower() for w in udhr.words(fileids=lang)]:
            ret.append(lang)
    return ret

print(word_in_langs("a"))
print(word_in_langs("the"))
print(word_in_langs("human"))
```

```python
def polysemy(word_type):
    word_type = list(word_type)
    return sum([len(w.lemmas()) for w in word_type]) / len(word_type)

print(polysemy(nouns))
print(polysemy(wn.all_synsets("v")))
print(polysemy(wn.all_synsets("s")))
print(polysemy(wn.all_synsets("r")))
```

```python
from nltk.corpus import wordnet as wn

word_pairs = "car-automobile, gem-jewel, journey-voyage, boy-lad, coast-shore, asylum-madhouse, magician-wizard, midday-noon, furnace-stove, food-fruit, bird-cock, bird-crane, tool-implement, brother-monk, lad-brother, crane-implement, journey-car, monk-oracle, cemetery-woodland, food-rooster, coast-hill, forest-graveyard, shore-woodland, monk-slave, coast-forest, lad-wizard, chord-smile, glass-magician, rooster-voyage, noon-string"
word_pairs = word_pairs.split(", ")

scores = {}

for pair in word_pairs:
    words = pair.split("-")
    first = wn.synsets(words[0])[0]
    second = wn.synsets(words[1])[0]
    scores[pair] = first.path_similarity(second)

print("Path similarity:")
print(sorted(scores, key=scores.get, reverse=True))

```
